```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("data/tableau_import.csv")
```

```python
df2 = df.sample(n=10000)
df2.loc[df2['non_white_population']<0, :] = np.nan
df2 = df2.dropna()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df2['non_white_population'], df2['partisanship_change'])
plt.xlabel("Non-White Population (fraction)", fontsize=20)
plt.ylabel("Partisanship Change", fontsize=20)
plt.savefig('bimodal.png')
```

```python
def plot_interactions_w_offset(df, var, k):
    pg = sns.PairGrid(df, y_vars=var,x_vars = [x for x in df.columns if x != var][k:k+10], height=4)
    pg.map(plt.scatter)
    plt.show()

def plot_interactions(df, var):
    for k in range(((len(df.columns)-1)//10)+1):
        plot_interactions_w_offset(df.sample(n=10000), var, k*10)
```

```python
plot_interactions(df, 'partisanship_change')
```
