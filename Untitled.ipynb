```python
remove_neg_list = ['density', 'total_population', 'white_population', 'hispanic_population', 'foreign_born_population',
                'median_household_income', 'mean_household_income', 'poverty_population', 'total_housing_units', 'units_paying_rent',
                'agriculture_industry', 'information_industry', 'financial_industry', 'entertainment_industry',
                'government_employee', 'armed_forces_population', 'degree_associate', 'degree_bachelor', 'degree_higher', 'veteran_population',
                'non_institutionalized_population', 'political_influence_2010', 'political_interest_2010']
for col in remove_neg_list:
    df.loc[df[col]<0, col] = np.nan

df = df.dropna()
```

```python
def prepare_data(data, nVecs):
    from numpy import nan, inf
    data = data.copy()
    data["white_population"] = 1 - (data['white_population']/100)
    data.rename(columns={"white_population":"non_white_population"}, inplace=True)
    data['hispanic_population'] /= data['total_population']
    data['foreign_born_population'] /= data['total_population']
    data['mean_household_income'] -= data['median_household_income']
    data['poverty_population'] /= 100
    data['units_paying_rent'] /= data['total_housing_units']
    data['total_housing_units'] /= data['total_population']
    data['unemployed_population'] /= data['total_population']
    data['agriculture_industry'] /= data['working']
    data['information_industry'] /= data['working']
    data['manufacturing_industry'] /= data['working']
    data['financial_industry'] /= data['working']
    data['entertainment_industry'] /= data['working']
    data['government_employee'] /= data['working']
    data['degree'] = (data['degree_associate'] + data['degree_bachelor'] + data['degree_higher']) / data['working']
    data['armed_forces_population'] /= data['working']
    data['veteran_population'] /= data['total_population']
    data['no_insurance'] /= data['non_institutionalized_population']
    data['public_insurance'] /= data['non_institutionalized_population']
    data['incarcerated'] = 1 - (data['non_institutionalized_population']/data['total_population'])
    
    keep_list = ['density', 'total_population', 'non_white_population', 'hispanic_population', 'foreign_born_population',
                'median_household_income', 'mean_household_income', 'poverty_population', 'total_housing_units', 'units_paying_rent',
                'agriculture_industry', 'information_industry', 'financial_industry', 'entertainment_industry',
                'government_employee', 'armed_forces_population', 'degree', 'veteran_population',
                'incarcerated', 'political_influence_2010', 'political_interest_2010', 'partisanship_influence_2010', 'partisanship_2010']
    keep_list += [x for x in data.columns if 'eigvec' in x and int(x.split("eigvec_")[1]) > 159-nVecs]
    keep_list += ['political_interest_2019', 'partisanship_2019']
    data = data[keep_list]
    data['to_delete'] = data['political_interest_2019'] + data['political_interest_2010']
    data['to_delete'] = data['to_delete'].replace(0, nan)
    data = data.dropna().drop(columns=['to_delete'])
    data['partisanship_change'] = (data['partisanship_2019'] - data['partisanship_2010'])/(data['political_interest_2019']+data['political_interest_2010'])
    return data
```

```python
from tensorflow import keras
from tensorflow.keras import layers
model = tf.keras.Sequential([
    layers.Dense(X_data.shape[1]*2, activation='elu', input_shape=[X_data.shape[1]], use_bias=True, kernel_initializer = tf.keras.initializers.GlorotUniform()),
    layers.Dropout(0.35),
    layers.Dense(X_data.shape[1], activation='elu', use_bias = True, kernel_initializer = tf.keras.initializers.GlorotUniform()),
    layers.Dropout(0.25),
    layers.Dense(64, activation='elu', use_bias=True, kernel_initializer = tf.keras.initializers.GlorotUniform()),
    layers.Dense(16, activation='linear', use_bias=True),
    layers.Dense(1, activation='linear', use_bias = False)
])
optimizer = tf.keras.optimizers.Adam()
model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

```python
EPOCHS = 500
model.fit(
  X_data, Y_data_part,
  epochs=EPOCHS, validation_split = 0.2, verbose=1)
```
